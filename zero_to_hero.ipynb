```python
# Password Retry
# Hardcode a correct password.
# Use a while loop that allows 3 attempts.
# If the user enters the correct password → print “Access Granted”
# Else after 3 wrong attempts → print “Access Denied”.

p = 'Hello@123'
a = 2
i = 1
while i<=3:
  n= input("Enter the password : ")
  if n == p:
    print("Access Granted")
    break
  else:
    print("Wrong Password, Attempt left : ", a)
  i+=1
  a -= 1
else:
  print('Access Denied')
```

```python
# E-commerce Orders Validation
orders = {
    "o1": {"amount": "100", "status": "paid"},
    "o2": {"amount": "-50", "status": "paid"},
    "o3": {"amount": "200", "status": "failed"},
    "o4": {"amount": "invalid", "status": "paid"}
}
# Tasks:

# Calculate total valid paid revenue
# Ignore invalid and negative amounts
# Find highest valid paid amount
rev = 0
max_amount = 0
for i in orders.values():
  if i['status'] == 'paid':
    if i['amount'].isdigit():
      rev += int(i['amount'])
  if i['amount'].isdigit():
    if int(i['amount']) > max_amount:
      max_amount = int(i['amount'])
    print('Ignore invalid and negative amounts:',int(i['amount']))
print('total valid paid revenue:',rev)
print('highest valid paid amount:',max_amount)
```
