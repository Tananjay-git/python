```python
# Simple Calculator
# Take two numbers and an operator (+, -, *, /) as hardcoded values
# Based on the operator, perform the corresponding arithmetic operation.
# Use if–elif–else.

a = float(input("Enter First Number : "))
b = float(input("Enter Second Number : "))

c = input("Enter the Operation : '+', '-', '*', '/': ")

if c == '+':
  print(a+b)
elif c == '-':
  print(a-b)
elif c == '*':
  print(a*b)
elif c == '/':
  print(a/b)
else:
  print("Choose the correct operator")
```

```python
# Discount System
# A store gives:
# 10% discount if total ≥ 500
# 20% if total ≥ 1000
# 0% otherwise
# Print the final bill after applying discount.
# Use nested if and arithmetic operators.

n = float(input("Enter the amount : "))
if n >= 500:
  print("Final Billing Amount: "+str(abs(n-(n*0.1))))
elif n >= 1000:
  print("Final Billing Amount: "+str(abs(n-(n*0.2))))
else:
  print("Final Billing  Amount: "+ str(abs(n)))
```

```python
# Password Retry
# Hardcode a correct password.
# Use a while loop that allows 3 attempts.
# If the user enters the correct password → print “Access Granted”
# Else after 3 wrong attempts → print “Access Denied”.

p = 'Hello@123'
a = 2
i = 1
while i<=3:
  n= input("Enter the password : ")
  if n == p:
    print("Access Granted")
    break
  else:
    print("Wrong Password, Attempt left : ", a)
  i+=1
  a -= 1
else:
  print('Access Denied')
```

```python
# Calculate Total Revenue from Order Amounts Stored as Strings

# Each order amount is stored as a string.
# Convert each value to float and compute the total revenue. Ignore empty string values.
# Example Input: ["12.5", "20", "", "7.5"]
# Expected Output: 40.0

li =  ["12.5", "20", "", "7.5"]
su = 0.0
for i in range(len(li)):
  if li[i] == '':
    continue
  a = float(li[i])
  su += a
print(su)
```

```python
# Sanitize Whitespace From Log Entries

# Given a list of strings where each entry may have leading/trailing spaces,
# create a clean list that removes only leading/trailing whitespace using loops and string slicing (no .strip()).
# Example Input: [" OK", "FAIL ", " TIMEOUT "]
# Expected Output: ["OK","FAIL","TIMEOUT"]
l = [" OK", "FAIL ", " TIMEOUT "]

for i in range(len(l)):
  l[i] = l[i].strip()
print(l)
```
